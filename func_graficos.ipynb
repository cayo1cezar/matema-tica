```python
import numpy as np
import matplotlib.pyplot as plt
```



Plotando uma função de 1º GRAU



```python
x = np.linspace(-5,5,10)
```

```python
y = 3*x+7
```

```python
plt.plot(x,y)
plt.grid()
plt.show()
```

Plotando uma função do 2º GRAU

```python
x2 = np.linspace(-15,15,50)
```

```python
y2 = 2*x2**2+6*x2+8
```

```python
plt.plot(x2,y2)
plt.grid()
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

Plotando uma função Exponencial 

```python
x3 = np.linspace(-5,20,50)

y3 = 2**x3
```

```python
plt.plot(x3,y3)
plt.grid()
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

Plotando uma função logarítmica (inversa da exponencial)

```python
x4 = np.linspace(1,20,50)

y4 = np.log2(x4)

plt.plot(x4,y4)
plt.grid()
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

Plotando uma função trigonométrica 

```python
#Domínio 

xt = np.linspace(-2*np.pi,2*np.pi,100)

#Funções 

ys = np.sin(xt)

yc = np.cos(xt)

yt = np.tan(xt)
```

```python
plt.plot(xt,ys)
plt.grid()
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
plt.plot(xt,yc)
plt.grid()
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
plt.plot(xt,yt)
plt.grid()
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```
